# 🎬 NLP Movie Recommender: A Beginner's Walkthrough

Welcome! This notebook explains how we use **Natural Language Processing (NLP)** to build a movie recommendation engine from scratch.

## 1. Importing Libraries

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import os
```

## 2. Loading the Data
We use the **Enriched Movies** dataset which contains plots, actors, and directors fetched from the OMDb API.

```python
df = pd.read_csv('../Dataset/enriched_movies.csv')
df.head(2)
```

## 3. How the 'AI' reads the movie
We combine all the text (Plot, Genre, Director) into one long string. This gives the AI a 'profile' of the movie.

```python
df['content'] = df['Plot'].fillna('') + " " + df['Genre'].fillna('') + " " + df['Director'].fillna('')
print("Sample Content:", df['content'].iloc[0][:100])
```

## 4. TF-IDF Vectorization
Computers don't understand words, they understand numbers. **TF-IDF** converts our text into a mathematical matrix.

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['content'])
print("Matrix Shape:", tfidf_matrix.shape)
```

## 5. Finding Similarity
We use **Cosine Similarity** to find movies that are 'closest' to each other in the mathematical space.

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
print("Similarity calculated!")
```
